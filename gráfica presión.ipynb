```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import trapz, simps


#Ejemplo

T = 2
m = 0.804652


# Create the vectors X and Y
x = np.array(range(20))
y = (np.exp(1-(np.sqrt(x**2 + m**2))/T)*x**4)/(np.sqrt(x**2 + m**2))

# Create the plot
plt.plot(x,y)

# Show the plot
plt.show()

#Area under curve
# Compute the area using the composite trapezoidal rule.
area = trapz(y, dx=1)
print("area =", area)

# Compute the area using the composite Simpson's rule.
area = simps(y, dx=1)
print("area =", area)
```
