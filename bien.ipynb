```python
from scipy.integrate import quad
import numpy as np 
import scipy.integrate as integrate
import scipy.special as special


t = 4

#Constantes
Tc = 139.57
A  = 3/4
B  = (9.4/np.sqrt(27))
G  = B*(1.000001 - 1/t)**(0.1) #Acoplamiento efectivo
mg = np.sqrt(A*((t*Tc)**2)*(G**2)) #masa gluón
b  = 0.2
d  = 0.1
v  = 16
Aq = 1/3
mp = 139.57 #masa del pion
m1 = 2.2 #quark up
m2 = 4.2 #quark down
m3 = 95 #quark strange
mq =  np.sqrt((m1/mp)**2 + Aq*((t*Tc)**2)*(G**2)) + np.sqrt((m2/mp)**2 + Aq*((t*Tc)**2)*(G**2)) + np.sqrt((m3/mp)**2 + Aq*((t*Tc)**2)*(G**2))


#Bosones

#Derivada de C(T)
C  = (1.03)*(1.02 - 1/t)**(0.2)
d1 = ((t*Tc)**2)*C*b*((1.02 - 1/t)**(-1))

#derivada de G(T)
d5 = G*d*((t*Tc)**2)*((1.000001 - 1/t)**(-1))

#derivada de m(T) para gluón
d4 = ((t*Tc)*G + G*((t*Tc)**2)*d5)*((np.sqrt((1/3)*((t*Tc)**2)*(G**2)))**(-1))

def Eg(x):                                   
    return np.sqrt(x**2 + mg**2) 
                                    
#derivada de Eg
def d3(x): 
    return (mg/Eg(x))*(d4)


#derivada de fB
def f1(x):
    return 1/(np.exp((Eg(x))/(t*Tc))-1)
def d2(x):
    return (-1/(np.exp((Eg(x))/(t*Tc))-1)**2)*(np.exp((Eg(x))/(t*Tc)))*(((1/t*Tc)*(d3(x)))-((1/t*Tc)**2)*(Eg(x)))


#derivada total bosonica
def DB(x):
    return (v/6*(np.pi)**2)*(x**4)*(d1*(f1(x)) + d2(x)*C)*(1/(Eg(x))) - C*(f1(x))*(1/(Eg(x))**2)*(d3(x))
                                    
#Fermiones

#Derivada de C(T)
C  = (1.03)*(1.02 - 1/t)**(0.2)
d11 = ((t*Tc)**2)*C*b*((1.02 - 1/t)**(-1))

#derivada de G(T)
d55 = G*d*((t*Tc)**2)*((1.000001 - 1/t)**(-1))
                                         

#derivada de m(T) para quarks
d44 = ((t*Tc)*G + G*((t*Tc)**2)*d55)*(np.sqrt((1/3)*((t*Tc)**2)*(G**2))**(-1))

def ED(x):
     return np.sqrt(x**2 + mq**2)                                       
                                      
#derivada de ED; Esta integral es una suma de tres integrales debido a los tres tipos de quarks
def d33(x):
    return (mq/(ED(x)))*(d44)
                                      
                               
#derivada de fD
def fD(x):
    return 1/(np.exp((ED(x))/(t*Tc))+1)
                                      
def d22(x): 
    return (-1/(np.exp((ED(x))/(t*Tc))-1)**2)*(np.exp((ED(x))/(t*Tc)))*(((1/t*Tc)*(d33(x)))-((1/t*Tc)**2)*(ED(x)))


#derivada total bosonica
def DD(x): 
    return (x**4)*(2/(np.pi)**2)*((d11*(fD(x)) + (d22(x))*C)*(1/(ED(x))) - C*(fD(x))*(1/((ED(x)))**2)*(d33(x)))

#Derivada de B(T): DP - S
def f(x): 
    return DB(x) + DD(x) - (v/2*(np.pi**2)*(t*Tc))*(x**2)*C*(f1(x))*((4/3)*(x**2) + (mg)**2)*(1/(Eg(x))) + (6/((t*Tc)*(np.pi)**2))*(x**2)*C*(fD(x))*((4/3)*(x**2) + (mq)**2)*(1/(ED(x)))


y, err = quad(f, 0, 20)

print(y)

                               
```

```python
#Integrando las dos derivadas totales
def f(x): 
    return DB(x) + DD(x) 

z, err = quad(f, 0, 20)

print(z)                                                                        
          

#Entropía
def k(x):
    return (v/2*(np.pi**2)*(t*Tc))*(x**2)*C*(f1(x))*((4/3)*(x**2) + (mg)**2)*(1/(Eg(x))) + (6/((t*Tc)*(np.pi)**2))*(x**2)*C*(fD(x))*((4/3)*(x**2) + (mq)**2)*(1/(ED(x)))

s, err = quad(k, 0, 20)

print(s)

#Calculando derivada de B(T)
print(z-s)
```
